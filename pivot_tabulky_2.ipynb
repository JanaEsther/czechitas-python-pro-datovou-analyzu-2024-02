```python
import pandas as pd
import numpy as np
import statistics
food_nutrient = pd.read_csv("food_nutrient.csv")
branded_food = pd.read_csv("branded_food.csv")
food = pd.concat([pd.read_csv("food_sample_100.csv"), pd.read_csv("food_other.csv")], ignore_index=True)
food_merged = pd.merge(food, food_nutrient, on="fdc_id")
food_merged = pd.merge(food_merged, branded_food, on="fdc_id")
food_merged = food_merged[food_merged["name"].isin(["Protein", "Sodium, Na", "Total lipid (fat)", "Carbohydrate, by difference", "Cholesterol"])]
```

```python
pd.crosstab(food_merged["branded_food_category"], food_merged["name"], food_merged["amount"], aggfunc=statistics.mean)
```

```python
food_merged = food_merged.drop_duplicates(subset=["fdc_id", "name"])
food_nutrient_pivot = pd.pivot(food_merged, index="fdc_id", columns="name", values="amount")
food_nutrient_pivot = pd.merge(food_nutrient_pivot, branded_food, on="fdc_id")
bins = [0, 20, 100, float('inf')]
labels = ['Low Cholesterol', 'Moderate Cholesterol', 'High Cholesterol']
food_nutrient_pivot['Cholesterol Category'] = pd.cut(food_nutrient_pivot['Cholesterol'], bins=bins, labels=labels)
pd.crosstab(food_nutrient_pivot["branded_food_category"], food_nutrient_pivot["Cholesterol Category"], normalize ="index")
```

```python
pd.crosstab(food_nutrient_pivot["branded_food_category"], food_nutrient_pivot["Cholesterol Category"])
```
