```python
import pandas as pd
```

```python
marathon = pd.read_csv("half_marathon.csv")
marathon.head(15)
```

```python
marathon["Cas_shift"] = marathon["Cas"].shift(periods=1)
marathon["Rok zavodu"] = 2020;
marathon
```

```python
marathon["Cas"] = pd.to_date(marathon["Cas"])
marathon["Cas2"] = pd.to_date(marathon["Cas"]).shift(-1)
```

```python
marathon = marathon[marathon["Rok zavodu"] == 2020].copy()
marathon["diff"] = marathon ["Cas"] - marathon["Cas2"] 
marathon["diff"] = marathon["diff"].dt.total_seconds()
marathon["diff"].mean()
```
